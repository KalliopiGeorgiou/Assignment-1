The director of graduate admission at a large university is analyzing the relationship between scores on the math portion of Graduate Record Examination (GRE) and subsequent performace in graduate school, as measured by student`s avrage (GPA). the accompaanying data file shows GPA and GRE of 24 students.

```python
import numpy as np
import pandas as pd
gpaDf =  pd.read_excel("https://www.dropbox.com/scl/fi/24jfn2u4rsn6e45bvzkrs/GPA.xlsx?rlkey=9ecqtc90bmntcsucdy8gzyt24&dl=1")
gpaDf
```

    a) (2points) Find regression model equation for the model: GPA = b0 + B1 GRE

```python
# https://www.statsmodels.org/stable/index.html
import statsmodels.api as sm
# Your answer
GPARegModel1 = sm.OLS(
  gpaDf["GPA"],
  sm.add_constant(gpaDf[["GRE"]])
)
GpaRegModel1Fit = GPARegModel1.fit()
print(GpaRegModel1Fit.summary())
```

```python
# Based on the regression results, the R squared is 0.27 which indicates that our model does not do a very good job at predicting GPA using only the GRE score.
# An R squared of 0.27 means that only 27% of the variance in GPA is explained by the GRE score. 
# This means that there are probably other factors that have a bigger influence on GPA than the GRE score.
```

    b) (1 point) what is a student`s prediction GPA if he/she scored 710 on math portion of GRE

```python
# Your answer
constant = 0.4256
b1 = 0.0041
GRE_score = 710
GPA = constant + b1*GRE_score
print(f'The estimated GPA of a student that scored 710 on math portion of GRE is: {GPA:.2f}')
```

10. The maintenance manager at a trucking company wants to build a regression model to forcast the time unyil the first engine overhaul(Time in years) based on four predictor variables for 25 trucks:
    1. anual miles driven (Miles in 1000s)
    2. avrage load weight (Load in tons)
    3. average driving speed (Speed in mph)
    4. oil change interval (Oil in 1000s of miles).

```python
import numpy as np
import pandas as pd
engineDf =  pd.read_excel("https://www.dropbox.com/scl/fi/dcoz9yw3f8yywtzy82f4z/Engine.xlsx?rlkey=n53hfjjsrddywktksj156jra5&dl=1")
engineDf.head()
```

```python
engineDf.shape
```

```python
engineDf.describe()
```

    a) (2 point)find model regression equation (use all four predictor variables)

```python
# your answer
# https://www.statsmodels.org/stable/index.html
import statsmodels.api as sm

EngineOverhaulRegModel1 = sm.OLS(
  engineDf["Time"],
  sm.add_constant(engineDf[["Miles","Load","Speed","Oil"]])
)
EngineOverhaulRegModel1Fit = EngineOverhaulRegModel1.fit()
print(EngineOverhaulRegModel1Fit.summary())
```

```python
# Our model regression equation using all four predictor variables is : Time = 13.43 -0.089*Miles - 0.073*Load + 0.0048*Speed - 0.0297*Oil
```

    b) (1point) Predict the time before the first engine overhaul for particular truck driven 60000 miles per year with an average load of 22 tons, an average driving speed of 57 mph, and 18000 miles between oil changes.

```python
# Your answer
Miles = 60
Load = 22
Speed = 57
Oil = 18
Time = 13.43 - 0.089*Miles - 0.073*Load + 0.0048*Speed - 0.0297*Oil
print(f'The estimated time before the first engine overhaul for the particular track is: {Time:.2f} years')
```
